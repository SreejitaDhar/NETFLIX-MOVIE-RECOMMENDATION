## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing dataset

```python
dataset = pd.read_csv('Netflix Movie Recommendation.csv', header = None)                             ##change in line
transactions = []

for i in range(0, 7466):                                                                             ##change in line
  transactions.append([str(dataset.values[i, j]) for j in range(0, 20)])
```

## Eclat Training on Dataset

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualizing

### Raw Results

```python
results = list(rules)
results
```

### Proper Display

```python
def inspect(results):
    movie_1         = [tuple(result[2][0][0])[0] for result in results]                                                          ##change in line
    movie_2         = [tuple(result[2][0][1])[0] for result in results]                                                          ##change in line
    supports    = [result[1] for result in results]

    ##change in line

    return list(zip(movie_1, movie_2, supports))                                                                                 ##change in line
DataFrame_intelligence = pd.DataFrame(inspect(results), columns = ['Movie 1', 'Movie 2', 'Support'])                             ##change in line
```
